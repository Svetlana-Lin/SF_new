### 1.2 Вынесение повторяющихся функций


```python
# Задаем функцию преобразования столбца, содержащего несколько вариантов ответов, результат которой:
# 1). все варианты ответов кодируются числами (1,2,3...)
# 2). пропущенный ответ заменяется на 0
# 3). добавляются новые столбцы с двоичной кодировкой для каждого варианта ответа (0-нет, 1-есть),
# 4). создается словарь соответствия текстового варианта ответа и численного кода

def columns_transform(df, col_lable, type_answ=0): 

    result_dict = {}
    
    i=1
    for item in df[col_lable].value_counts().index.tolist():
        if type_answ==1: 
            if 'Да,' in item:
                result_dict[item] = 1
            elif 'Возможно' in item:
                result_dict[item] = 2
            elif 'Скорее' in item:
                result_dict[item] = 3   
            elif 'Нет,' in item:
                result_dict[item] = 4    
        else:    
            result_dict[item] = i
            i+=1
            

    df[col_lable] = df[col_lable].map(result_dict)
    codes = sorted(list(result_dict.values()))
        
    for_rename = {}
    for code in codes:
        for_rename[code] = col_lable+'__'+str(code)
        
    df[codes] = (pd.get_dummies(df[col_lable]))
    df = df.rename(columns=for_rename)
    
    df[col_lable] = df[col_lable].fillna(0)
    df[col_lable] = df[col_lable].astype (int)
    
    if (len(df[df[col_lable]==0])>0):
        result_dict['Не ответили/вопрос не задавался'] = 0    

    
    return df, result_dict 
```

```python
# Задаем функцию для построения графиков динамики в Plotly

def graph_plotly(df, col_lables, title):
    fig = go.Figure(layout={'title': title})
    for col in col_lables:
        fig.add_trace(go.Scatter(x=df.date, y=df[col], name=col))

    iplot(fig, show_link=False) # для Jupyter 
    # fig.show(renderer="colab") # для Colab
```

```python
# Задаем функцию для вывода на экран таблиц среднего и медианы в группах 

def tabl_avg_med(df, col_group, col_var, new_col):
    
    df = df[df[col_var]>0]
    
    df_stat = df.groupby(col_group)[[col_group, col_var]].agg({col_group:'count', col_var:['mean', 'median']}).round()
    df_stat.columns = df_stat.columns.droplevel(0)
    df_stat.sort_values(by=['count'], ascending=False)
    df_stat.columns = new_col
    
    return df_stat
```

```python
# Задаем функцию для вывода на экран таблиц распределения вопросов относительно пола

def pivot_tabl_count(df, col):
    pivot = df.pivot_table(values = ['Ваш возраст:'],
                           index = ['Ваш пол:'],
                           columns = [col],
                           aggfunc = 'count',  
                           margins = True,
                           fill_value=0).div(len(df.index)).mul(100) # зададим вывод в %
    pivot.columns = pivot.columns.droplevel(0)
    return pivot
```

#### 3.5.1 Между полом респондентов и его намерением вести учет существует взаимосязь

Н0: Между полом респондента и его намерением вести учет нет взаимосвязи (группы не зависимы)  
Н1: Между полом респондента и его намерением вести учет есть зависимость (группы зависимы) 

**Выберем подходящий статистический тест**

Для выбора нужного теста воспользуемся алгоритмом выбора теста. Для этого ответим на следующие вопросы:
* Какой тип у признака «Пол»? — Категориальный.
* Сколько сравниваемых групп? — Две.
* Проверяется независимость групп? — да.

Для проверки нашей гипотезы можно использовать критерий ${\chi}^2$ (хи-квадрат).

**Проведём тест**

```python
# составляем таблицу сопряжённости
cross_table = pd.crosstab(df1['gender'], df1['q10'])

cross_table
```

```python
# проводим тест
_, p, _, _ = stats.chi2_contingency(cross_table)
decision_hypothesis(p)
```

Важно помнить, что нормальность будет проявляться только при достаточно больших частотах. В статистике принято считать, что **общее количество наблюдений** (сумма частот) должна быть **не менее 50** и ожидаемая частота **в каждой группе должна быть не менее 5**. Только в этом случае величина, показанная выше, имеет стандартное нормальное распределение. 

```python
cross_table['Положительный'] = cross_table['Возможно'] + cross_table['Хочу']
cross_table['Отрицательный'] = cross_table['Не буду'] + cross_table['Не хочу']
cross_table_ = cross_table.drop(['Возможно', 'Не буду', 'Хочу', 'Не хочу'], axis = 1)
cross_table_
```

```python
# проводим тест
_, p, _, _ = stats.chi2_contingency(cross_table_)
decision_hypothesis(p)
```

*p-значение в статистически правльно сгруппированных данных гораздо больше первого варианта, и не смотря на то, что в этом случае результат не изменился, все-таки не соблюдение наполнености групп может дать ложную оценку.*

**Вывод**: Между полом респондента и его намерением вести учет нет взаимосвязи (группы не зависимы)
